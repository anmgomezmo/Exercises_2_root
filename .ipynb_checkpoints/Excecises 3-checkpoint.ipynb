```c++
#include <iostream>
#include <string>
#include <stdio.h>
```

```c++
TFile *file = TFile::Open("../Tracks_Clusters.root");
TTree *tree = (TTree*) file->Get("JetRecoTree");
tree->Print();
```

# Exercise 1

```c++
float evtw = -1;
vector<float> *reco_R4_pt;
vector<float> *truth_R4_pt;

tree->SetBranchAddress("EventWeight",&evtw);
tree->SetBranchAddress("RecoJets_R4_pt",&reco_R4_pt);
tree->SetBranchAddress("TruthJets_R4_pt",&truth_R4_pt);

TCanvas *canvas = new TCanvas("Canvas","",800,600);

TH1F *hist_leadreco_pt = new TH1F("Lead Reco-Jet","Leading Jet pt;pt (GeV);Events",50,0,200);
TH1F *hist_truth_pt = new TH1F("Truth Jet","Truth & Reco Jet pt;pt (GeV);Events",50,0,200);
```

```c++
int nentries, nbytes, i;
nentries = (Int_t)tree->GetEntries();

for(i=0;i<nentries;i++){
    nbytes = tree->GetEntry(i);
    if(reco_R4_pt->size()>0){
        hist_leadreco_pt->Fill(reco_R4_pt->at(0)/1000.);
        for(int j=0;j<reco_R4_pt->size();j++){
            hist_truth_pt->Fill(truth_R4_pt->at(0)/1000.);
        }
    }
}

std::cout << "Done! " <<std::endl;
```

```c++
hist_truth_pt->SetMarkerStyle(20);
hist_truth_pt->SetMarkerColor(kRed);
hist_truth_pt->Draw("");
hist_leadreco_pt->SetMarkerStyle(21);
hist_leadreco_pt->Draw("same");
canvas->SetLogy();
canvas->Draw();
```

```c++
for(i=0;i<nentries;i++){
    nbytes = tree->GetEntry(i);
    if(reco_R4_pt->size()>0){
        hist_leadreco_pt->Fill(reco_R4_pt->at(0)/1000.,evtw);
        for(int j=0;j<reco_R4_pt->size();j++){
            hist_truth_pt->Fill(truth_R4_pt->at(0)/1000.,evtw);
        }
    }
}

std::cout << "Done! " <<std::endl;
```

```c++
hist_truth_pt->SetMarkerStyle(20);
hist_truth_pt->SetMarkerColor(kRed);
hist_truth_pt->Draw("");
hist_leadreco_pt->SetMarkerStyle(21);
hist_leadreco_pt->Draw("same");
canvas->SetLogy();
canvas->Draw();
```

# Excercise 3

```c++
UInt_t npv = -1;
tree->SetBranchAddress("NPV", &npv);

float mu = -1;
tree->SetBranchAddress("mu_average", &mu);

TProfile *prof_recoR4pt_npv = new TProfile("Profile Reco-jet pT vs. NPV",";NPV; jet pT",50,1,50, 0, 200);
TProfile *prof_recoR10pt_npv = new TProfile("Profile Reco-jet pT vs. NPV",";NPV; jet pT",50,1,50, 0, 200);
TProfile *prof_truthR4pt_npv = new TProfile("Profile Truth-jet pT vs. NPV",";NPV; jet pT",50,1,50, 0, 200);
TProfile *prof_truthR10pt_npv = new TProfile("Profile Truth-jet pT vs. NPV",";NPV; jet pT",50,1,50, 0, 200);
```

```c++
for (i = 0; i < nentries; i++)
{
    nbytes = tree->GetEntry(i);

    if(reco_R4_pt->size()!=0 && reco_R4_pt->at(0)>20000.){
        for(int j=0; j<reco_R4_pt->size(); j++){
        prof_recoR4pt_npv->Fill(reco_R4_pt->at(j)/1000.,npv,evtw);  
        }
    }
    
    if(reco_R10_pt->size()!=0 && reco_R10_pt->at(0)>20000.){
        for(int j=0; j<reco_R4_pt->size(); j++){
        prof_recoR10pt_npv->Fill(reco_R10_pt->at(j)/1000.,npv,evtw);
        }
    }
    
    if(truth_R4_pt->size()!=0 && truth_R4_pt->at(0)>20000.){
        for(int j=0; j<reco_R4_pt->size(); j++){
        prof_truthR4pt_npv->Fill(truth_R4_pt->at(j)/1000.,npv,evtw);
        }
    }
    
    if(truth_R10_pt->size()!=0 && truth_R10_pt->at(0)>20000.){
        for(int j=0; j<reco_R4_pt->size(); j++){
        prof_truthR10pt_npv->Fill(truth_R10_pt->at(j)/1000.,npv,evtw);
        }
    }      
        
}

std::cout << "Done!" << std::endl;
```

```c++
prof_recoR4pt_npv->Draw("");
canvas->Draw();
```

```c++
prof_recoR10pt_npv->Draw("");
canvas->Draw();
```

```c++
prof_truthR4pt_npv->Draw("");
canvas->Draw();
```
