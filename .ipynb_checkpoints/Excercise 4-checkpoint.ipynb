```c++
#include <iostream>
#include <string>
#include <stdio.h> 
```

```c++
TFile *file = TFile::Open("../Tracks_Clusters.root");
TTree *tree = (TTree*) file->Get("JetRecoTree");
tree->Print();
```

# Exercise 4

```c++
float evtw = -1;
vector<float> *reco_R4_pt;
vector<float> *truth_R4_pt;
vector<float> *reco_R4_jvf;
UInt_t npv = -1;
float mu = -1;

tree->SetBranchAddress("mu_average", &mu);
tree->SetBranchAddress("NPV", &npv);
tree->SetBranchAddress("EventWeight",&evtw);
tree->SetBranchAddress("RecoJets_R4_pt",&reco_R4_pt);
tree->SetBranchAddress("TruthJets_R4_pt",&truth_R4_pt);
tree->SetBranchAddress("RecoJets_R4_jvf",&reco_R4_jvf);

TH1F *hist_leadrecopt_cut_jvf = new TH1F("Lead Reco-jet cut JVF","Leading reco-jet JVF; pT (GeV);Events",20,0,1000);
TH1F *hist_leadtruth_pt = new TH1F("Lead Truth-jet JVF","Leading truth jet;pT (GeV);Events",50,0,1000);
TH1F *hist_leadreco_pt = new TH1F("Lead Reco-jet cut JVF","Leading reco jet;pT (GeV);Events",50,0,1000);

TCanvas *canvas = new TCanvas("Canvas","",800,600);
```

```c++
int nentries, nbytes, i;
nentries = (Int_t)tree->GetEntries();

for (i = 0; i < nentries; i++)
{
    nbytes = tree->GetEntry(i);
    
    if(truth_R4_pt->size()>0){
        hist_leadtruth_pt->Fill(truth_R4_pt->at(0)/1000.,evtw);
    }
    
    if(reco_R4_pt->size()>0){
        hist_leadreco_pt->Fill(reco_R4_pt->at(0)/1000.,evtw);
    }
    
    
    if(reco_R4_jvf->at(0)>0.5 || reco_R4_jvf->at(0)<-0.5){
        if(reco_R4_pt->size()>0){
            hist_leadrecopt_cut_jvf->Fill(reco_R4_pt->at(0)/1000.,evtw);
        }
    }   
}
   
std::cout << "Done!" << std::endl;
```

```c++
hist_leadrecopt_cut_jvf->SetMarkerStyle(22);
hist_leadrecopt_cut_jvf->SetMarkerColor(kGreen);
hist_leadrecopt_cut_jvf->Draw("");
canvas->SetLogy();
canvas->Draw();
```

```c++
hist_leadtruth_pt->SetMarkerStyle(20);
hist_leadtruth_pt->SetMarkerColor(kRed);
hist_leadtruth_pt->Draw("");
hist_leadreco_pt->SetMarkerStyle(21);
hist_leadreco_pt->SetMarkerColor(kBlue);
hist_leadreco_pt->Draw("same");
hist_leadrecopt_cut_jvf->SetMarkerStyle(22);
hist_leadrecopt_cut_jvf->SetMarkerColor(kGreen);
hist_leadrecopt_cut_jvf->Draw("same");
canvas->SetLogy();
canvas->Draw();
canvas->SaveAs("./Images/lead_truth-reco-jvf_cut.png");
```

```c++
TProfile *prof_leadtruth_pt_npv = new TProfile("Profile Truth-jet pT vs. NPV",";NPV; jet pT",50,1,50,0,200);
TProfile *prof_leadreco_pt_npv = new TProfile("Profile Reco-jet pT vs. NPV",";NPV; jet pT",50,1,50,0,200);
TProfile *prof_leadreco_pt_jvf_npv = new TProfile("Profile Reco-jet cut jvf pT vs. NPV",";NPV; jet pT",50,1,50,0,200);

TProfile *prof_leadtruth_pt_mu = new TProfile("Profile Truth-jet pT vs. mu",";mu; jet pT",50,1,100,0,200);
TProfile *prof_leadreco_pt_mu = new TProfile("Profile Reco-jet pT vs. mu",";mu; jet pT",50,1,100,0,200);
TProfile *prof_leadreco_pt_jvf_mu = new TProfile("Profile Reco-jet cut jvf pT vs. mu",";mu; jet pT",50,1,100,0,200);

TCanvas *canvas = new TCanvas("Canvas","",800,600);
```

```c++
for (i = 0; i < nentries; i++)
{
    nbytes = tree->GetEntry(i);

    if(reco_R4_pt->size()!=0){
        for(int j=0; j<reco_R4_pt->size(); j++){
            prof_leadreco_pt_npv->Fill(reco_R4_pt->at(j)/1000.,npv,evtw);
            prof_leadreco_pt_mu->Fill(reco_R4_pt->at(j)/1000.,mu,evtw);
        }
    }
    
    if(truth_R4_pt->size()!=0){
        for(int j=0; j<truth_R4_pt->size(); j++){
            prof_leadtruth_pt_npv->Fill(truth_R4_pt->at(j)/1000.,npv,evtw);
            prof_leadtruth_pt_mu->Fill(truth_R4_pt->at(j)/1000.,mu,evtw);
        }
    }
    
    if(reco_R4_jvf->at(0)>0.5 || reco_R4_jvf->at(0)<-0.5){
        if(reco_R4_pt->size()!=0){
            for(int j=0; j<reco_R4_pt->size(); j++){
                prof_leadreco_pt_jvf_npv->Fill(reco_R4_pt->at(j)/1000.,npv,evtw);
                prof_leadreco_pt_jvf_mu->Fill(reco_R4_pt->at(j)/1000.,mu,evtw);
            }
        }
    }

}

std::cout << "Done!" << std::endl;
```

```c++
prof_leadreco_pt_jvf_npv->SetMarkerStyle(22);
prof_leadreco_pt_jvf_npv->SetMarkerColor(kGreen);
prof_leadreco_pt_jvf_npv->SetStats(0);
prof_leadreco_pt_jvf_npv->Draw("");
prof_leadtruth_pt_npv->SetMarkerStyle(20);
prof_leadtruth_pt_npv->SetMarkerColor(kRed);
prof_leadtruth_pt_npv->Draw("same");
prof_leadreco_pt_npv->SetMarkerStyle(21);
prof_leadreco_pt_npv->SetMarkerColor(kBlue);
prof_leadreco_pt_npv->Draw("same");
canvas->Draw();
canvas->SaveAs("./Images/npv_truth-reco-jvf_cut.png");
```

```c++
prof_leadreco_pt_jvf_mu->SetMarkerStyle(22);
prof_leadreco_pt_jvf_mu->SetMarkerColor(kGreen);
prof_leadreco_pt_jvf_mu->SetStats(0);
prof_leadreco_pt_jvf_mu->Draw("");
prof_leadtruth_pt_mu->SetMarkerStyle(20);
prof_leadtruth_pt_mu->SetMarkerColor(kRed);
prof_leadtruth_pt_mu->Draw("same");
prof_leadreco_pt_mu->SetMarkerStyle(21);
prof_leadreco_pt_mu->SetMarkerColor(kBlue);
prof_leadreco_pt_mu->Draw("same");
canvas->Draw();
canvas->SaveAs("./Images/mu_truth-reco-jvf_cut.png");
```
